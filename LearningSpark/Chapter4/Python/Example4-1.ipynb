## Ejemplo 4.1

#### Realizamos los imports

```
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

#### Creamos un conversor a Fecha en formato UDF

```
def to_date_format_udf(d_str):
    l = [char for char in d_str]
    return "".join(l[0:2]) + "/" +  "".join(l[2:4]) + " " + " " +"".join(l[4:6]) + ":" + "".join(l[6:])

to_date_format_udf("02190925")
```

#### Registramos el UDF

```
spark.udf.register("to_date_format_udf", to_date_format_udf, StringType())
```

#### Leemos el fichero

```
df = (spark.read.format("csv")
     .schema("date STRING, delay INT, distance INT, origin STRING, destination STRING")
     .option("header", "true")
     .option("path", "/databricks-datasets/learning-spark-v2/flights/departuredelays.csv")
     .load())

display(df)
```

#### Testeamos el UDF

```
df.selectExpr("to_date_format_udf(date) as data_format").show(10, truncate=False)
```

#### Creamos una vista temporal

```
df.createOrReplaceTempView("us_delay_flights_tbl")
```

#### Cacheamos la tabla para agilizar las consultas

```
%sql
CACHE TABLE us_delay_flights_tbl
```

#### Mostramos las 10 primeras líneas usando nuestro UDF

```
spark.sql("SELECT *, date, to_date_format_udf(date) AS date_fm FROM us_delay_flights_tbl").show(10, truncate=False)
```

```
spark.sql("SELECT COUNT(*) FROM us_delay_flights_tbl").show()
```

#### Mostramos los primeros 10 resultados de los vuelos cuya distancia sea mayor a 1000 ordenados de mayor a menor

```
spark.sql("SELECT distance, origin, destination FROM us_delay_flights_tbl WHERE distance > 1000 ORDER BY distance DESC").show(10, truncate=False)
```

#### Equivalencias de la sentencia anterior

```
df.select("distance", "origin", "destination").where(col("distance") > 1000).orderBy(desc("distance")).show(10, truncate=False)
```

```
df.select("distance", "origin", "destination").where("distance > 1000").orderBy("distance", ascending=False).show(10)
```

```
df.select("distance", "origin", "destination").where("distance > 1000").orderBy(desc("distance")).show(10)
```

#### Mostramos todos los vuelvos con 2 horas de retraso entre San Francisco y Chicago

```
spark.sql("""SELECT date, delay, origin, destination FROM us_delay_flights_tbl WHERE delay > 120 AND origin = 'SFO' AND destination = 'ORD' ORDER BY delay DESC""").show(10, truncate=False)
```

#### Mostramos todos los vuelos etiquetando según si no tienen retraso, tienen poco retraso, tienen retraso moderado o tienen mucho retraso independientemente de su destino

```
spark.sql("""SELECT delay, origin, destination, CASE 
    WHEN delay > 360 THEN 'Very Long Delays'
    WHEN delay > 120 AND delay < 360 THEN 'Long Delays'
    WHEN delay > 60 AND delay < 120 THEN 'Short Delays'
    WHEN delay > 0 AND delay < 60 THEN 'Torelable Delays'
    ELSE 'No delays'
  END AS Flight_Delays FROM us_delay_flights_tbl ORDER BY origin, delay DESC""").show(10, truncate=False)
```
