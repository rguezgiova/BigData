#### GlobalTempView vs TempView

TempView crea una referencia del DataFrame en uso y su tiempo de vida es la sesión en la que el DataFrame fue creado, mientras que GlobalTempView se comparten entre todas las sesiones y muere cuando todas las aplicaciónes se terminan.
