#### Leer el CSV del ejemplo del cap2 y obtener la estructura del schema dado por defecto.

```
%scala
import org.apache.spark.sql.functions._

val file = "/databricks-datasets/learning-spark-v2/mnm_dataset.csv"

val mnmDf =  spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(file)

mnmDf.schema
mnmDf.printSchema
```

#### Cuando se define un schema al definir un campo por ejemplo `StructField('Delay',FloatType(), True)` ¿qué significa el último parámetro Boolean?

El último parámetro significa si puede ser nulo o no

#### DataSet vs DataFrame (Scala). ¿En qué se diferencian a nivel de código?

En los DataFrames se pueden lanzar consultas SQL, en los DataSet además puede lanzar funciones. Adicionalmente en los DataSet es necesario definir una case class para definir el objeto.

#### Utilizando el mismo ejemplo utilizado en el capítulo para guardar en parquet y guardar los datos en los formatos:
  i) JSON <br>
  ii) CSV (dándole otro nombre para evitar sobrescribir el fichero origen) <br>
  iii) AVRO

```
%scala
mnmDf.write.format("json").mode("overwrite").save("/tmp/mnm-json")
```

```
%scala
mnmDf.write.format("csv").mode("overwrite").save("/tmp/mnm-csv")
```

```
%scala
mnmDf.write.format("avro").mode("overwrite").save("/tmp/mnm-avro")
```

#### Revisar al guardar los ficheros (p.e. json, csv, etc) el número de ficheros creados, revisar su contenido para comprender (constatar) como se guardan.
  i) ¿A qué se debe que hayan más de un fichero? <br>
  Spark crea un fichero por cada partición empleada. <br>
  
  ii) ¿Cómo obtener el número de particiones de un DataFrame?
  Se obtiene con la función **getNumPartitions**.

```
%scala
mnmDf.rdd.getNumPartitions
```

iii) ¿Qué formas existen para modificar el número de particiones de un DataFrame? <br>
Se utiliza la función **repartition(partitionNumber)** <br>

iiii) Llevar a cabo el ejemplo modificando el número de particiones a 1 y revisar de nuevo el/los ficheros guardados.

```
%scala
val repartitionMnmDf = mnmDf.repartition(2)
repartitionMnmDf.write.format("parquet").mode("overwrite").save("/tmp/mnm-parquet")
```
