## Ejemplo Line Count

#### Vemos la versión de Spark

```
%scala
spark.version
```

#### Definmimos el fichero y vemos las 10 primeras líneas

```
%scala
val strings = spark.read.text("/databricks-datasets/learning-spark-v2/SPARK_README.md")
strings.show(10, false)
```

#### Contamos las líneas del fichero

```
%scala
strings.count()
```

#### Filtramos las líneas que contengan la palabra "_Spark_"

```
%scala
val filtered = strings.filter($"value".contains("Spark"))
filtered.count
```
