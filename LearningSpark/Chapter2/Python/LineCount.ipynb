### Ejemplo Line Count

#### Vemos la versión de Spark

```
spark.version
```

#### Definimos el fichero y mostramos las 10 primeras líneas

```
strings = spark.read.text("/databricks-datasets/learning-spark-v2/SPARK_README.md")
strings.show(10, truncate=False)
```

#### Contamos las líneas del fichero

```
strings.count()
```

#### Filtramos las líneas que contengan la palabra _"Spark"_

```
filtered = strings.filter(strings.value.contains("Spark"))
filtered.count()
```
