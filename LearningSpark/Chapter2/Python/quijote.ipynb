## Ejemplo El Quijote

#### Definimos el fichero .txt

```
file = "/FileStore/shared_uploads/giovanni.rodriguez@bosonit.com/el_quijote.txt"
```

#### Leemos el fichero

```
quijoteDf = spark.read.text(file)
quijoteDf.persist()
```

#### Contamos las líneas totales

```
quijoteDf.count()
```

#### Mostramos las primeras 20 líneas

```
quijoteDf.show()
```

#### Mostramos las primeras 5 líneas

```
quijoteDf.show(5)
```

#### Mostramos las primeras 5 líneas con todo su contenido

```
quijoteDf.show(5, truncate=False)
```

#### Muestra las primeras 20 líneas con todo su contenido

```
quijoteDf.show(truncate=False)
```

#### Muestra la primera línea

```
quijoteDf.first()
```

#### Muestra la primera línea

```
quijoteDf.head()
```

#### Muestra las primeras 5 líneas

```
quijoteDf.head(5)
```

#### Muestra las primeras 5 líneas

```
quijoteDf.take(5)
```
